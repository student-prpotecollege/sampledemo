```python
#program 1: User Login Check

# Real-World Application: Authentication Systems

username = "admin"
password = "1234"

# Taking input from user
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking login condition
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# program 2: Pass / Fail Analyzer
# Real-World Application: Academic Evaluation Systems

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# program 3: Simple Data Cleaner
# Real-World Application: Data Preprocessing

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
# program 4: Message Length Analyzer
# Real-World Application: Text Validation System

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("-> This message is longer than 10 characters")
```

```python
# Program 5: Error Message Detector
# Real-World Application: Log Monitoring System

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
