```python
#Problem 1: Employee Performance Bonus Eligibility
# Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find maximum score manually
top_score = 0
for score in employees.values():
    if score > top_score:
        top_score = score

# Step 2: Find employees who have this score
bonus_list = []

for name in employees:
    if employees[name] == top_score:
        bonus_list.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(bonus_list), f"(Score: {top_score})")
```

```python
#Problem 2: Search Query Keyword Analysis
# Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation manually (simple way)
cleaned = ""
for ch in query:
    if ch.isalnum() or ch == " ":
        cleaned += ch

words = cleaned.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Print only repeated words
result = {}
for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#Problem 3: Sensor Data Validation
# Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_data = []

i = 0
while i < len(sensor_readings):
    value = sensor_readings[i]

    if value % 2 == 0:  # even values are valid
        valid_data.append((i, value))

    i += 1

print("Valid Sensor Readings (Hour, Value):")
print(valid_data)
```

```python
#Problem 4: Email Domain Usage Analysis
# Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_data = {}

for mail in emails:
    parts = mail.split("@")
    domain = parts[1]

    if domain in domain_data:
        domain_data[domain] += 1
    else:
        domain_data[domain] = 1

total = len(emails)

print("Email Domain Usage:")
for domain in domain_data:
    percent = (domain_data[domain] / total) * 100
    print(domain, ":", int(percent), "%")
```

```python
#Problem 5: Sales Spike Detection
# Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average manually
total_sales = 0
for value in sales:
    total_sales += value

avg = total_sales / len(sales)

limit = avg + (0.30 * avg)

print("Sales Spike Days:")
day_number = 1

for value in sales:
    if value > limit:
        print("Day", day_number, ":", value)
    day_number += 1
```

```python
#Problem 6: Duplicate User ID Detection
# Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_data = {}

for uid in user_ids:
    if uid in count_data:
        count_data[uid] += 1
    else:
        count_data[uid] = 1

print("Duplicate User IDs:")

for uid in count_data:
    if count_data[uid] > 1:
        print(uid, "→", count_data[uid], "times")
```
