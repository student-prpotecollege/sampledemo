```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
users = pd.read_json("users.json")
users.head()
```

```python
conn = sqlite3.connect(":memory:")

with open("restaurants.sql", "r", encoding="utf-8") as f:
    conn.executescript(f.read())

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
final_df = orders.merge(users, on="user_id", how="left") \
                 .merge(restaurants, on="restaurant_id", how="left")

final_df.head()
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)

```

```python
import pandas as pd

# Load the dataset
# Ensure 'final_food_delivery_dataset.csv' is in the same directory
df = pd.read_csv('final_food_delivery_dataset.csv')

# --- SECTION 1: Numerical Questions ---
print("--- SECTION 1: Numerical Answers ---")

# 1. Total orders by users with Gold membership
gold_orders = df[df['membership'] == 'Gold']
q1 = len(gold_orders)
print(f"1. Total Gold orders: {q1}")

# 2. Total revenue from orders in Hyderabad (rounded to nearest integer)
hyd_revenue = df[df['city'] == 'Hyderabad']['total_amount'].sum()
q2 = round(hyd_revenue)
print(f"2. Total Revenue from Hyderabad: {q2}")

# 3. Distinct users who placed at least one order
q3 = df['user_id'].nunique()
print(f"3. Distinct Users: {q3}")

# 4. Average order value for Gold members (rounded to 2 decimals)
avg_gold_val = gold_orders['total_amount'].mean()
q4 = round(avg_gold_val, 2)
print(f"4. Average Order Value (Gold): {q4}")

# 5. Orders placed for restaurants with rating >= 4.5
q5 = len(df[df['rating'] >= 4.5])
print(f"5. Orders with Rating >= 4.5: {q5}")

# 6. Orders in the top revenue city among Gold members only
# Group by city for Gold members and sum revenue
gold_city_revenue = gold_orders.groupby('city')['total_amount'].sum()
# Get the city with the max revenue
top_gold_city = gold_city_revenue.idxmax()
# Count orders in that city for Gold members
q6 = len(gold_orders[gold_orders['city'] == top_gold_city])
print(f"6. Orders in Top Revenue City (Gold Members) [{top_gold_city}]: {q6}")


# --- SECTION 2: Fill-in-the-Blank Answers ---
print("\n--- SECTION 2: Fill-in-the-Blank Answers ---")
print("1. The column used to join orders.csv and users.json is: user_id")
print("2. The dataset containing cuisine and rating information is stored in: JSON format")
print(f"3. The total number of rows in the final merged dataset is: {len(df)}")
print("4. If a user has no matching record in users.json, the merged values will be: NaN")
print("5. The Pandas function used to combine datasets based on a key is: merge")
print("6. The column membership in the final dataset originates from the: users.json file")
print("7. The join key used to combine orders data with restaurant details is: restaurant_id")
print("8. The column that helps identify the type of food served by a restaurant is: cuisine")
print("9. If a user places multiple orders, their personal details appear: multiple times")
```
